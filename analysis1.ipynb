```python
# filter and get just the needed columns
nextpred = df[(df['Type']=='FUNCTIONAL') & (df['FunctionName::PropertyName'] == 'OnChanged_i_NextPredicted::')][['Object Name', 'FunctionName::PropertyName','Message', 'Date  Hour']]
nextpred
```

### lets split the columns to lists and do ops on them

```python
timecol = list(nextpred.iloc[:,3])
timecol
```

```python
col1 = list(nextpred.iloc[:,0])
col1
```

```python
col1sep = [[x.split('_')[1],x.split('_')[2].split('.')[0] ] for x in col1]
col1sep
```

```python
col2 = list(nextpred.iloc[:,2])
col2
```

```python
# cutting out initial sentence to get just the XML content
col2xml = [x[42:] for x in col2]
col2xml
```

### XML to Dict/Json conversion

```python
col2json = [ xmltodict.parse(x, attr_prefix='') for x in col2xml]
col2json
```

```python
[ len(x['PREDICTEDs']['PREDICTED']) for x in col2json]
# shows that each entry has 3 predictions.
```

```python
# getting to the main part
col2json1 = [ x['PREDICTEDs']['PREDICTED']  for x in  col2json]
```

```python
json.dumps(col2json1)
```

### Looping through the json array and making a flattened Prediction Array
Each entry is carrying 3 predictions. Flattening that, so that in the result array there are 3 rows per original entry. Giving each prediction its own row.

```python
predictArray = []
situationList = set()
tripsList = set()
serviceList = set()
numberList = set()
for i,row in enumerate(col2json1):
    jsonRow = {'direction': col1sep[i][0], 'stop_id':col1sep[i][1], 'timestamp':timecol[i] }
    pred = [{},{},{}]
    
    for x in range(3):
        pred[x]['NUMBER'] = row[x]['NUMBER']
        pred[x]['PREDICTED_ARRIVAL'] = row[x].get('PREDICTED_ARRIVAL','')
        pred[x]['PREDICTED_DEPARTURE'] = row[x].get('PREDICTED_DEPARTURE','')
        pred[x]['SERVICE'] = row[x]['SERVICE']
        pred[x]['TRIP'] = row[x]['TRIP']
        pred[x]['SITUATION'] = row[x]['SITUATION']
        pred[x]['PREDICTED_DESTINATION_PLATFORM'] = row[x].get('PREDICTED_DESTINATION_PLATFORM','')
        pred[x].update(jsonRow)
        
        # looking up stations csv data to get the stop_id for PREDICTED_DESTINATION_PLATFORM
        testinternal = pred[x]['PREDICTED_DESTINATION_PLATFORM']
        pred[x]['pred_dest_stop_id'] = list(dfStations[(dfStations['down_id'] == testinternal) | (dfStations['up_id'] == testinternal) ]['stop_id'])[0]
        predictArray.append( pred[x] )
    
    situationList.update([row[0]['SITUATION'], row[1]['SITUATION'], row[2]['SITUATION']  ])
    tripsList.update([row[0]['TRIP'], row[1]['TRIP'], row[2]['TRIP']  ])
    serviceList.update([row[0]['SERVICE'], row[1]['SERVICE'], row[2]['SERVICE']  ])
    numberList.update([row[0]['NUMBER'], row[1]['NUMBER'], row[2]['NUMBER']  ])
```

```python
# finally, this is what one row of the prediction array looks like
json.dumps(predictArray[0])
```

### Saving Prediction array to a CSV

```python
dfPred = pd.DataFrame(predictArray)
dfPred.to_csv('output/nextPrediction.csv', index_label='sr')
```
